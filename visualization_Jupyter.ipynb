```python
# Extracting the relevant columns for the scatter plot
x_values = governance_df['governance_avg']
y_values = governance_df['happiness_score']

# Customize the plot style
plt.style.use('ggplot')  # Optional: You can choose from various styles like 'seaborn', 'ggplot', etc.

# Create scatter plot of happiness_score against governance_avg
plt.scatter(x_values, y_values, c='orange', s=50, alpha=0.7, edgecolors='black')

# Add titles and labels
plt.title('Happiness Score vs Governance Average')
plt.xlabel('Governance Average')
plt.ylabel('Happiness Score')

# fitting a linear regression line
m, b = np.polyfit(x_values, y_values, 1)

# adding the regression line to the scatter plot
plt.plot(x_values, m*x_values + b, color = "blue")


# Show the plot
plt.show()
```

```python
# Replace zeros with NaN
soc_df.replace(0, pd.NA, inplace=True)
```

```python
# Calculate the row-wise average
soc_df['soc_avg'] = soc_df[['average_working_hours', 'household_size', 'education']].mean(axis=1)

soc_df
```

```python
# Extracting the relevant columns for the scatter plot
x_values = soc_df['soc_avg']
y_values = soc_df['happiness_score']

# Customize the plot style
plt.style.use('ggplot')  # Optional: You can choose from various styles like 'seaborn', 'ggplot', etc.

# Create scatter plot of happiness_score against governance_avg
plt.scatter(x_values, y_values, c='orange', s=50, alpha=0.7, edgecolors='black')

# Add titles and labels
plt.title('Happiness Score vs Social Average')
plt.xlabel('Social Average')
plt.ylabel('Happiness Score')

# Show the plot
plt.show()
```

```python
soc_df['average_working_hours'] = pd.to_numeric(soc_df['average_working_hours'], errors='coerce')
soc_df['happiness_score'] = pd.to_numeric(soc_df['happiness_score'], errors='coerce')

# Remove rows where x_values or y_values contain NaN values
clean_data = soc_df.dropna(subset=['average_working_hours', 'happiness_score'])

# Recreate x_values and y_values from the cleaned data
x_values = clean_data['average_working_hours']
y_values = clean_data['happiness_score']

# Plot
plt.scatter(x_values, y_values, c='green', s=50, alpha=0.7, edgecolors='black')
plt.title('Happiness Score vs Average Working Hours')
plt.xlabel('Average Working Hours')
plt.ylabel('Happiness Score')

# fitting a linear regression line
m, b = np.polyfit(x_values, y_values, 1)

# adding the regression line to the scatter plot
plt.plot(x_values, m*x_values + b, color="red")

plt.show()
```

```python
#The trend suggests that countries or individuals with shorter working hours tend to report higher happiness levels. 
#This highlights the importance of maintaining a healthy work-life balance to improve well-being and avoid burnout.
```

```python
soc_df['household_size'] = pd.to_numeric(soc_df['household_size'], errors='coerce')
soc_df['happiness_score'] = pd.to_numeric(soc_df['happiness_score'], errors='coerce')

# Remove rows where x_values or y_values contain NaN values
clean_data = soc_df.dropna(subset=['household_size', 'happiness_score'])

# Recreate x_values and y_values from the cleaned data
x_values = clean_data['household_size']
y_values = clean_data['happiness_score']

# Plot
plt.scatter(x_values, y_values, c='orange', s=50, alpha=0.7, edgecolors='black')
plt.title('Happiness Score vs Household Size')
plt.xlabel('Household Size')
plt.ylabel('Happiness Score')


# fitting a linear regression line
m, b = np.polyfit(x_values, y_values, 1)

# adding the regression line to the scatter plot
plt.plot(x_values, m*x_values + b, color="blue")


plt.show()
```
