```python
import pandas as pd
from wgi_functions import drop_columns, rename_columns, get_country_code, reformat_country_names
import difflib
```

```python
# Load data from the World Governance Indicators dataset

voice_accountability = pd.read_excel("wgidataset2022.xlsx", sheet_name=0)

political_stability = pd.read_excel("wgidataset2022.xlsx", sheet_name=1)

government_effectiveness = pd.read_excel("wgidataset2022.xlsx", sheet_name=2)

regulatory_quality = pd.read_excel("wgidataset2022.xlsx", sheet_name=3)

rule_of_law = pd.read_excel("wgidataset2022.xlsx", sheet_name=4)

control_corruption = pd.read_excel("wgidataset2022.xlsx", sheet_name=5)
```

```python
voice_accountability = drop_columns(voice_accountability)
political_stability = drop_columns(political_stability)
government_effectiveness = drop_columns(government_effectiveness)
regulatory_quality = drop_columns(regulatory_quality)
rule_of_law = drop_columns(rule_of_law)
control_corruption = drop_columns(control_corruption)
```

```python
voice_accountability.Estimate = voice_accountability.Estimate.round(2)
voice_accountability.Rank = voice_accountability.Rank.round(2)
```

```python
political_stability.Estimate = political_stability.Estimate.round(2)
political_stability.Rank = political_stability.Rank.round(2)
```

```python
government_effectiveness.Estimate = government_effectiveness.Estimate.round(2)
government_effectiveness.Rank = government_effectiveness.Rank.round(2)
```

```python
regulatory_quality.Estimate = regulatory_quality.Estimate.round(2)
regulatory_quality.Rank = regulatory_quality.Rank.round(2)
```

```python
rule_of_law.Estimate = rule_of_law.Estimate.round(2)
rule_of_law.Rank = rule_of_law.Rank.round(2)
```

```python
control_corruption.Estimate = control_corruption.Estimate.round(2)
control_corruption.Rank = control_corruption.Rank.round(2)
```

```python
voice_accountability = rename_columns(voice_accountability)
political_stability = rename_columns(political_stability)
government_effectiveness = rename_columns(government_effectiveness)
regulatory_quality = rename_columns(regulatory_quality)
rule_of_law = rename_columns(rule_of_law)
control_corruption = rename_columns(control_corruption)
```

```python
governance_df = voice_accountability.merge(
    political_stability, on='country', how='outer', suffixes=('_voice', '_stability')
).merge(
    government_effectiveness, on='country', how='outer', suffixes=('', '_gov')
).merge(
    regulatory_quality, on='country', how='outer', suffixes=('', '_quality')
).merge(
    control_corruption, on='country', how='outer', suffixes=('', '_corruption')
).merge(
    rule_of_law, on='country', how='outer', suffixes=('', '_law')
)
```
