```python
country_mapping = {
    'Brunei Darussalam': 'Brunei',
    'Kyrgyz Republic': 'Kyrgyzstan',
    'Korea': 'South Korea',  # Assuming it should match South Korea
    'St. Lucia': 'Saint Lucia',
    'Marshall Islands': 'Marshall Islands',  # Assuming no change
    'West Bank and Gaza': 'Palestine',
    'Russian Federation': 'Russia',
    'Slovak Republic': 'Slovakia',
    'Eswatini': 'Swaziland',
    'Timor-Leste': 'East Timor',
    'Turkiye': 'Turkey',
    'Viet Nam': 'Vietnam'
}
```

```python
df['country'] = df['country'].replace(country_mapping)
```

```python
remaining_mismatches = df[~df['country'].isin(df_2['country'])]
print(remaining_mismatches['country'].unique())
```
