```python
import pandas as pd

file_path = '/Users/nancybaraka/Downloads/undesa_pd_2022_hh-size-composition (1).xlsx'
excel_data = pd.ExcelFile(file_path)

excel_data.sheet_names
```

```python
df = excel_data.parse('HH size and composition 2022')

df.head()
```

```python
df.columns
```

```python
df=df.iloc[4:, :].copy()
df
```

```python
df = df[['Country or area', 'Reference date (dd/mm/yyyy)', 'Average household size (number of members)']]
df
```

```python
df['Reference date (dd/mm/yyyy)'] = pd.to_datetime(df['Reference date (dd/mm/yyyy)'], errors='coerce')
df
```

```python
df['Reference date (dd/mm/yyyy)'] = df['Reference date (dd/mm/yyyy)'].dt.year
df
```

```python
df = df.sort_values('Reference date (dd/mm/yyyy)', ascending=False).groupby('Country or area').head(1)

df

```
